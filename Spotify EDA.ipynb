```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
# Optional: Remove non-ASCII characters from relevant columns
df['album'] = df['album'].str.encode('ascii', 'ignore').str.decode('ascii')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='album', y='popularity', color='teal', s=50)

plt.title('Album vs Popularity')
plt.xlabel('album')
plt.ylabel('Popularity')
plt.grid(True)
plt.tight_layout()
plt.show()
```
